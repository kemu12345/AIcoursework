**2.3 Outlier Detection**

**Detect outliers in HDI and GNI per Capita.**

```python
def find_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)
```

```python
sa_latest.loc[:, 'HDI_Outlier'] = find_outliers_iqr(sa_latest['hdi'])
sa_latest.loc[:, 'GNI_Outlier'] = find_outliers_iqr(sa_latest['gross_inc_percap'])

print(sa_latest[['country', 'HDI_Outlier', 'GNI_Outlier']].head())
```

**PROBLEM 3**

**3.1 Create Middle East and South Asia Subsets**

**Define Middle East countries, filter the 2020–2022 dataset to create regional subsets, and save as CSV.**

```python
# Load the dataset created in Problem 1B
hdi_1B = pd.read_csv("HDI_problem1B.csv")

# Define lists from assignment text
south_asia = ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]
middle_east = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan", "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar", "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"]

# Filter Subsets
hdi_sa_2020_2022 = hdi_1B[hdi_1B['country'].isin(south_asia)].copy()
hdi_me_2020_2022 = hdi_1B[hdi_1B['country'].isin(middle_east)].copy()

# Save as CSV
hdi_sa_2020_2022.to_csv("HDI_SouthAsia_2020_2022.csv", index=False)
hdi_me_2020_2022.to_csv("HDI_MiddleEast_2020_2022.csv", index=False)
```

**3.2 Descriptive Statistics**

**Compute mean and standard deviation of HDI for both regions and identify the better performer.**

```python
# Compute stats
sa_stats = hdi_sa_2020_2022['hdi'].agg(['mean', 'std'])
me_stats = hdi_me_2020_2022['hdi'].agg(['mean', 'std'])

print("South Asia Stats:\n", sa_stats)
print("\nMiddle East Stats:\n", me_stats)

# Identification logic for report
if me_stats['mean'] > sa_stats['mean']:
    print("\nResult: Middle East performs better on average.")
else:
    print("\nResult: South Asia performs better on average.")
```

**3.3 Top and Bottom Performers**

**Identify top 3 and bottom 3 countries in each region and compare using a bar chart.**

```python
# Get latest year for ranking
sa_latest = hdi_sa_2020_2022[hdi_sa_2020_2022['year'] == 2022]
me_latest = hdi_me_2020_2022[hdi_me_2020_2022['year'] == 2022]

performers = pd.concat([
    sa_latest.nlargest(3, 'hdi'), sa_latest.nsmallest(3, 'hdi'),
    me_latest.nlargest(3, 'hdi'), me_latest.nsmallest(3, 'hdi')
])

me_top = me_latest.nlargest(3, 'hdi')
me_bottom = me_latest.nsmallest(3, 'hdi')


plt.figure(figsize=(12, 6))
sns.barplot(data=performers, x='hdi', y='country', hue='country', palette='coolwarm', legend=False)
plt.title("Top and Bottom Performers: South Asia vs Middle East (2022)")
plt.xlabel("HDI Value")
plt.show()
```

**3.4 Metric Comparisons**

**Compare Gender Development, Life Expectancy, and GNI per Capita across regions using grouped bar charts.**

```python
# Add region label for grouping
hdi_sa_2020_2022['region'] = 'South Asia'
hdi_me_2020_2022['region'] = 'Middle East'
combined_regions = pd.concat([hdi_sa_2020_2022, hdi_me_2020_2022])

# Plotting metrics (Example: Gender Development)
metrics = ['gender_development', 'life_expectancy', 'gross_inc_percap']

for metric in metrics:
    plt.figure()
    sns.barplot(data=combined_regions, x='region', y=metric, errorbar=None)
    plt.title(f"Comparison of {metric} across Regions")
    plt.show()
```

**3.5 HDI Disparity**

**Compute Range and Coefficient of Variation (CV) for both regions.**

```python
def compute_disparity(df):
    hdi_range = df['hdi'].max() - df['hdi'].min()
    cv = (df['hdi'].std() / df['hdi'].mean()) * 100
    return hdi_range, cv

sa_range, sa_cv = compute_disparity(hdi_sa_2020_2022)
me_range, me_cv = compute_disparity(hdi_me_2020_2022)

print(f"South Asia: Range={sa_range:.3f}, CV={sa_cv:.2f}%")
print(f"Middle East: Range={me_range:.3f}, CV={me_cv:.2f}%")
```
